````python
%%R -o dailyaqhi
aqhi <- readRDS('aqhi-cleaned.rds')
aqhi$hourlyAQHI <- apply(aqhi[,3:17], 1, mean, na.rm=TRUE)
require(plyr)
dailyaqhi <- ddply(aqhi, .(Date), summarize, meanAQHI = mean(hourlyAQHI), medianAQHI = median(hourlyAQHI))
````

````python
dailyaqhi
````

````python
plot(dailyaqhi[1,:])
plot(dailyaqhi[2,:])
# Great tip of "rotation" from
#    http://stackoverflow.com/questions/3100985/plot-with-custom-text-for-x-axis-points
xticks(arange(len(dailyaqhi[1,:])), dailyaqhi[0,:], rotation=90)
legend(['meanAQHI', 'medianAQHI'])
title('daily AQHI')
````

````python
meanDailyAQHI = dailyaqhi[1,:].astype('float')
````

````python
%%R -o meanDailyVIS
janWeather <- read.csv('jan_weather.csv')
meanDailyVIS <- janWeather$reduced.vis
````

````python
scatter(meanDailyAQHI, meanDailyVIS)
````

````python
print meanDailyAQHI
print meanDailyVIS
````

````python
%%R -i meanDailyAQHI,meanDailyVIS -o lineCoeff
print(meanDailyAQHI)
print(meanDailyVIS)
model <- lm(meanDailyVIS~meanDailyAQHI)
print(summary(model))
plot(model)
lineCoeff <- coef(lm(meanDailyVIS~meanDailyAQHI))
````

````python
scatter(meanDailyAQHI, meanDailyVIS)
_x = arange(3,8,0.1)
plot(_x, _x * lineCoeff[1] + lineCoeff[0], 'r', linewidth=3)
````
